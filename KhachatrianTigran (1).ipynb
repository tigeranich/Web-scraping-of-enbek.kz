```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

## Основной код 

```python
salary = []
region = []
experience = []
time = []
education = []
```

```python
url = 'https://enbek.kz/ru/search/vac?prof=Программист-аналитик&page=1'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')
items = soup.find_all('div', class_='item-list')

for n, i in enumerate(items, start=1):
    #зарплата
    itemSalary = i.find('div', class_='price').text.strip()
    salary.append(itemSalary)
    #регион
    itemRegion = i.find('li', class_='location d-flex align-items-center me-lg-3').text
    region.append(itemRegion)
    #стаж
    itemExpirience = i.find('li', class_='experience d-flex align-items-center').text
    experience.append(itemExpirience)
    #сколько работать
    itemTime = i.find('li', class_='time d-flex align-items-center me-lg-3').text
    time.append(itemTime)
    #требуемое образование
    itemEducation = i.find('li', class_='education d-flex align-items-center').text
    education.append(itemEducation)
    print(f'{n}: Регион: {itemRegion}, Зарплата: {itemSalary}')


pages = soup.find('ul', class_='pagination')
urls = []
links = pages.find_all('a')
links = links[2:]

for link in links:
    pageNum = int(link.text) if link.text.isdigit() else None
    if pageNum != None:
        hrefval = link.get('href')
        urls.append(hrefval)

k = 10        
a = list(range(34))[2:]
z = []
for i in a:
    z.append('&page='+str(i))


for slug in z:
    newUrl = url.replace('&page=1', slug)
    response = requests.get(newUrl)
    soup = BeautifulSoup(response.text, 'lxml')
    items = soup.find_all('div', class_='item-list')
    for n, i in enumerate(items, start=1):
        #зарплата
        try:
            itemSalary = i.find('div', class_='price').text.strip()
            salary.append(itemSalary)
        except:
            continue
        #регион
        itemRegion = i.find('li', class_='location d-flex align-items-center me-lg-3').text
        region.append(itemRegion)
        #стаж
        itemExpirience = i.find('li', class_='experience d-flex align-items-center').text
        experience.append(itemExpirience)
        #сколько работать
        itemTime = i.find('li', class_='time d-flex align-items-center me-lg-3').text
        time.append(itemTime)
        #требуемое образование
        itemEducation = i.find('li', class_='education d-flex align-items-center').text
        education.append(itemEducation)
        k += 1
        print(f'{k}: Регион: {itemRegion}, Зарплата: {itemSalary}')
```

```python
df = pd.DataFrame()
df['salary'] = salary
df['region'] = region
df['experience'] = experience
df['time'] = time
df['education'] = education
df
```

```python
df.to_csv('KhachatrianTigran.csv')
```
